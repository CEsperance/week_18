# 1. What is a neural network? What are the general steps required to build a neural network?

A neural network is a corrective feedback loop that rewards weights that support correct guesses and punishes weights for incorrect guesses. Neural Networks cluster unlabeled data based on similarities and classifies the data when it has a labeled dataset to train on. It is composed of several layers which are made of nodes (neurons). The nodes are where computation happens.
A node takes input from the data, combines it with a set of weights (coeefficients) thta amplify or dampen that specific input.
- the significance assigned to the featuress is in regard to what the algorithm is trying to learn (which feature is most helpful in classifying data without error, or the least amount of error).
- the input weight products are summned and then passed through a node's activation function to determine if and to what extent that signal should progress further through the network to affect the final outcome. If it passes through, the "neuron" has been activated. 
- the weighted inputs result in a guess that the neuron takes and compared it to a "ground_truth" about the data (did it get this right?)
- the difference between the network's guess and the ground truth is its error. The error is measured by the network and walks the error back over the model (backward propogation), adjusting the weights.

Below are three pseudo mathematical formulas I found in an artical that helped my understanding:
https://wiki.pathmind.com/neural-network#:~:text=Neural%20networks%20are%20a%20set,labeling%20or%20clustering%20raw%20input.
1. input * weight = guess 
(input enters the network and the weights map that input to a set of guesses that network makes at the end)
2. acutal output - guess = error
(the network takes the guess and compares it a ground-truth about the data)
3. error * weight's contribution to error = adjustment
(network measures that error and walks it back over the model adjusting the weights)

From launchcode
A neural network is a machine learning algorithm with the training data being the input to the input layer and the predicted value the value at the output layer.

Parameters:
Each connection from one neuron to another has an associated weight w. Each neuron, except the input layer which just holds the input value, also has an extra weight we call the bias weight, b. During feed-forward our input gets transformed by weight multiplications and additions at each layer, the output of each neuron can also get transformed by the application of what's called an activation function.

Gradient descent:
Learning in neural networks consists of tuning the weights or parameters to give the desired output. One way of achieving this is by using the famous gradient descent algorithm, and applying weight updates incrementally via a process known as back-propagation

## Feature Selection

a large set of features can cause overfitting and slow computing
use Pearson's R to remove highly correlated variables

```python
corr = heart_df.corr()
corr
```

```python
#selection highly correlated features
def correlated(df, threshold):
    col_corr = set()
    corr_matrix = df.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
cor_features = correlated(heart_df, 0.7)
len(set(cor_features))
```

```python
cor_features
```

```python
heart_df.drop(cor_features, axis=1, inplace = True)
heart_df
```

Why Keras?

- good for beginners and experts

Keras allows you to build models in two different ways; using either the Functional API or the Sequential API. We will focus on the Sequential API. This is a simple, yet very powerful way of building neural networks that will get you covered for most use cases. With the sequential API you're essentially building a model as a stack of layers. You can start with an input layer, add a couple of hidden layers, and finally end the model by adding an output layer


Keras Building a Model steps
1. specify architecture - number of layers, number of nodes in each layers, activation function for each layer
2. Compile the model - specify the loss function and details on how optimization works
3. Fit the model - cycle of back-propogation and optimization of model weights with your data 
4. Predict

# 5.Create a neural network using pytorch to predict the same result as question 3.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

X2 = X.values
y2 = y.values

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size = 0.3, random_state=42, stratify=y)

# #Standardize
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#create tensors = matrices 
X_train_t = torch.FloatTensor(X_train) 
X_test_t = torch.FloatTensor(X_test)

y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)

print(X_train)
print(y_train)
```

```python
#artificial neural network
class ANN_Model(nn.Module):
    def __init__(self, input_features=169,hidden1=20,hidden2=20,out_features=2):
        super().__init__() 
        self.layer_1_connection = nn.Linear(input_features, hidden1)
        self.layer_2_connection = nn.Linear(hidden1, hidden2)
        self.out = nn.Linear(hidden2, out_features)
        
    def forward(self, x):
        #apply activation functions
        x = F.relu(self.layer_1_connection(x))
        x = F.relu(self.layer_2_connection(x))
        x = self.out(x)
        return x
```

```python
torch.manual_seed(24)

#create instance of model
ann = ANN_Model()
```

```python
#loss function
loss_function = nn.CrossEntropyLoss()

#optimizer
optimizer = torch.optim.Adamax(ann.parameters(),lr=0.01)
```

```python
#run model through multiple epochs/iterations
final_loss = []
n_epochs = 500
for epoch in range(n_epochs):
    y_pred = ann.forward(X_train)
    loss = loss_function(y_pred, y_train)
    final_loss.append(loss)
    
    if epoch % 10 == 1:
        print(f'Epoch number: {epoch} with loss: {loss}')
        
    #zero the gradient before running backwards propagation
    optimizer.zero_grad() 
    loss.backward()
    #perform one optimization step each epoch
    optimizer.step() 
```
